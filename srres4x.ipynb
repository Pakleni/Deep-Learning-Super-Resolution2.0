```python
from dlsr import *
from tensorflow import keras

helpers.config(True)
```

```python
# general
image_size = 96
batch_size = 128
epochs = 500

n = 1e-4
```

```python
# setup training data
training_data = helpers.get_training_data(
    image_size=image_size, batch_size=batch_size, repeat_count=20, scale=4
)
```

```python
# instantiate model
srres = models.generator(B=16, scale=4)
loss_fn = keras.losses.MeanSquaredError()

# create a history object
history = helpers.History([
    "loss",
    "val_loss",
    "accuracy",
    "val_accuracy",
])
```

```python
train(
    model=srres,
    training_data=training_data,
    epochs=epochs,
    n=n,
    loss_fn=loss_fn,
    history=history,
)
```

```python
srres.save("./results/saved-models/srres4x.h5")
```

```python
history.plot({
    "loss": "Loss",
    "val_loss": "Validation Loss"
})
history.plot({
    "accuracy": "Accuracy",
    "val_accuracy": "Validation Accuracy"
})
```
